```python
from cobra import Model,Reaction,Metabolite
```

```python
glu_model=Model("glu_model")
```

## Add Reaction

```python
v0=Reaction('v0')
v0.name='v0'
v0.lower_bound=1
v0.upper_bound=1
```

```python
v1=Reaction('v1')
v1.name='v1'
v1.lower_bound=0
v1.pper_bound=1000
```

```python
v2=Reaction('v2')
v2.name='v2'
v2.lower_bound=0
v2.pper_bound=1000
```

```python
M=Reaction('M')
M.name='M'
M.lower_bound=0
M.pper_bound=1000
```

## Add Metabolites

```python
glu=Metabolite('glu',compartment='c')
AA=Metabolite('AA',compartment='c')
BioMass=Metabolite('BioMass',compartment='c')
```

```python
v0.add_metabolites({glu:1})
```

```python
v1.add_metabolites({glu:-1,AA:1})
```

```python
v2.add_metabolites({AA:-9.09,BioMass:1})
```

```python
M.add_metabolites({BioMass:-1})
```

```python
glu_model.add_reactions([v0,v1,v2,M])


glu_model.objective='M'


glu_model.optimize()
```
