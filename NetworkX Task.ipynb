##  read Edge List from file and Create a graph

```python
import matplotlib.pyplot as plt
```

```python
G=nx.read_edgelist('edges_List.txt',create_using=nx.Graph(),nodetype=int)
```

```python
nx.info(G)
```

```python
nx.draw(G,with_labels = True)
```

```python
list(G)
```

```python
G.edges()
```

```python
dict(G.degree())
```

```python
degree_sequence = sorted((d for n, d in G.degree()), reverse=True)
```

```python
degree_sequence
```

```python
nx.to_numpy_matrix(G)
```

```python
def plot_degree_dist(G):
    degrees = [G.degree(n) for n in G.nodes()]
    plt.xlabel("Degree", fontsize=20)
    plt.ylabel("Frequency", fontsize=20)
    plt.hist(degrees)
    plt.show()

plot_degree_dist(G)
```
